```python
import pandas as pd
```

```python
custleavingDataSet = pd.read_csv('churn.csv')
```

```python
custleavingDataSet.head()
```

```python
custleavingDataSet.describe()
```

```python
custleavingDataSet.hist(['CreditScore', 'Age', 'Balance'])
```

```python
custleavingDataSet.corr()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encode_CL_G = OrdinalEncoder()
```

```python
encode_CL_GR = OrdinalEncoder()
```

```python
custleavingDataSet['Geography_code'] = encode_CL_G.fit_transform(custleavingDataSet[['Geography']])
```

```python
custleavingDataSet['Gender_code'] = encode_CL_GR.fit_transform(custleavingDataSet[['Gender']])
```

```python
custleavingDataSet.head()
```

```python
custleavingDataSet.drop(columns=['Gender', 'Geography','Surname'], inplace=True)
```

```python
custleavingDataSet.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
custleaving_train, custleaving_test = train_test_split(custleavingDataSet, test_size=0.2)
```

```python
custleaving_train_X = custleaving_train.loc[:, custleaving_train.columns != 'Exited']
```

```python
custleaving_train_Y = custleaving_train['Exited']
```

```python
custleaving_train_Y.head()
```

```python
custleaving_train_X.head()
```

```python
custleaving_test_X = custleaving_test.loc[:, custleaving_test.columns != 'Exited']
```

```python
custleaving_test_Y = custleaving_test['Exited']
```

```python
custleaving_test_Y.head()
```

```python
custleaving_test_X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
custleavingModel = RandomForestClassifier(max_depth=2, random_state=0)
```

```python
custleavingModel.fit(custleaving_train_X, custleaving_train_Y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
custleavingprediction_Y = custleavingModel.predict(custleaving_test_X)
```

```python
custleavingprediction_Y
```

```python
accuracy_score(custleaving_test_Y,custleavingprediction_Y)
```
